```python
# Bibliotecas
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
```

```python
# Leitura do arquivo
gdf = gpd.read_file('/content/drive/MyDrive/GeoTech/DF_domicilios.gpkg')
```

```python
# Criar o gráfico hexbin
fig, ax = plt.subplots(figsize=(10, 10))
hb = ax.hexbin(gdf.geometry.x, gdf.geometry.y, gridsize=20, cmap='RdYlGn_r', alpha=0.5, mincnt=1)

# Adicionar o basemap
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik, crs=gdf.crs)

# Adicionar a barra de cores
cb = plt.colorbar(hb, ax=ax, shrink=0.5)
cb.set_label('Densidade de Pontos')
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```
