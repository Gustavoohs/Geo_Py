```python
# Bibliootecas
import geopandas as gpd
```

```python
# Leitura de arquivos
limite = gpd.read_file('DF.gpkg')
estacoes = gpd.read_file('estacoes_buffer.gpkg')
```

```python
# Reprojetando estações para coincidir com a camada limites
estacoes = estacoes.to_crs(limite.crs)
```

```python
# Visualizando estações
estacoes.plot()
```

```python
limite.plot()
```

```python
# Cálculo da diferença simétrica
diff = estacoes.dissolve().symmetric_difference(limite)
```

```python
# Visualizando resultado
diff.plot()
```
